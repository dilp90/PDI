# **Filtro Pasa Bajo (Suavizado) en Radiografías de Tórax**

**Objetivo**: Aplicar un filtro de suavizado para reducir el ruido en imágenes de radiografías de tórax.

Dataset: Chest X-Ray Images (Pneumonia) https://www.kaggle.com/datasets/paultimothymooney/chest-xray-pneumonia

Instrucciones:



1.   Carga del dataset: Descarga el dataset desde Kaggle y selecciona una imagen de radiografía de tórax.
2.   Aplicación del filtro pasa bajo:
*   Utiliza un filtro de suavizado, como el filtro de promedio (kernel de 5x5), para reducir el ruido en la imagen.
3.  Visualización de resultados:
* Muestra la imagen original y la imagen suavizada para comparar los efectos del filtro.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen en escala de grises
img = cv2.imread('/content/pneumonia_1.jpeg', cv2.IMREAD_GRAYSCALE)

# Definir el kernel de suavizado (filtro de promedio 5x5)
kernel = np.ones((7,7), np.float32) / 25

# Aplicar el filtro pasa bajo
img_suavizada = cv2.filter2D(img, -1, kernel)

# Mostrar la imagen original y la suavizada
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.title('Imagen Original')
plt.subplot(1,2,2), plt.imshow(img_suavizada, cmap='gray'), plt.title('Imagen Suavizada')
plt.show()
```

Para aumentar o disminuir el efecto del suavizado de imagen, se puede aumentar el tamaño del kernel.

# **Filtro Pasa Alto (Detección de Bordes) en Imágenes de Células Sanguíneas**

**Objetivo**: Aplicar un filtro para resaltar los bordes en imágenes de células sanguíneas.

**Dataset**: Blood Cell Images (https://www.kaggle.com/datasets/paultimothymooney/blood-cells)

Instrucciones:

1. Carga del dataset: Descarga el dataset desde Kaggle y selecciona una imagen de células sanguíneas.

2. Aplicación del filtro pasa alto:

* Utiliza un *filtro Laplaciano* para resaltar los bordes en la imagen. El filtro Laplaciano calcula la diferencia entre un píxel y sus vecinos cercanos en la imagen. Cuando un píxel está rodeado de valores de intensidad similares (áreas planas), el valor de la segunda derivada es cercano a cero. Sin embargo, cuando hay un cambio abrupto en la intensidad (es decir, un borde), la segunda derivada se vuelve significativa, lo que indica la presencia de un borde.
Desventaja: es sensible al ruido de la imagen.

3. Visualización de resultados:

* Muestra la imagen original y la imagen con los bordes resaltados para analizar los efectos del filtro.

```python
# Cargar la imagen en escala de grises
img = cv2.imread('/content/BloodImage_00001.jpg', cv2.IMREAD_GRAYSCALE)

# Definir el kernel Laplaciano (filtro pasa alto)
kernel = np.array([[0, -1, 0],
                   [-1, 4, -1],
                   [0, -1, 0]])

# Aplicar el filtro pasa alto
img_bordes = cv2.filter2D(img, -1, kernel)

# Mostrar la imagen original y la imagen con bordes resaltados
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.title('Imagen Original')
plt.subplot(1,2,2), plt.imshow(img_bordes, cmap='gray'), plt.title('Bordes Resaltados')
plt.show()
```
